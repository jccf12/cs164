```python
import numpy as np
import pandas as pd
import cvxpy as cp
```

```python
n = 5
A = np.zeros([n,n])
for j in range(5):
    A[0] = [np.random.random()*(1+int(x==0)) for x in range(5)]
    A[0] = A[0]/np.sum(A[0])

for i in range(1,5):
    A[i] =[np.random.random()*3]+[np.random.random()*(3-int((abs(i-j)+1)/2)) for j in range(1,5)]
    A[i] = A[i]/np.sum(A[i])
    
A
```

```python
O = np.array([42,17,13,20,8])
x0 = np.array([30,20,20,15,15])
B = np.zeros([n**2,n])
for i in range(n**2):
    for j in range(n):
        a = int(i/n)
        b = (i)%n
        if j == a:
            if b != a:
                B[i][j] = -1
        if j == b:
            if b != a:
                B[i][j] = 1
```

```python
def get_x_t(t):
    s = 0
    xt = x0
    while t>s:
        xt = np.dot(xt,A)
        s+=1
    return xt
```

```python
T = 3
x = [get_x_t(t) for t in range(T)]
```

```python
def get_x_t_B(t,ut):
    s = 0
    xt = x0
    while t>s:
        xt = np.dot(xt,A)+np.dot(ut,B)
        s+=1
    return xt
```

```python
u = cp.Variable(n**2*T, integer=True)
x = cp.Variable()
S= np.identity(n)
x1 = [get_x_t_B(t,u[n**2*t:n**2*t+n**2]) for t in range(T)]
objective = cp.Minimize(
    np.sum([cp.sum_squares(O-np.round(A@x[i])-u[n**2*i:n**2*i+n**2]@B) for i in range(T-1)])
    +cp.quad_form(x[T-1], S)
)
constraints = [
               np.ones(n**2*T)@u <= 100,
                u >= 0
               ]

prob = cp.Problem(objective,
                 constraints)
prob.solve()
```

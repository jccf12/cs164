```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f0(x):
    return x**2+1

#f_1(x) = x -4 <= 0
#f_2(x) = -x + 2 <= 0

def phi(x):
    return -(np.log(-x+4)+np.log(x-2))
```

```python
X = np.linspace(1,5,50)
X_dom = np.linspace(2.0001,3.9999,500)
yf0 = [f0(x) for x in X]

plt.figure(figsize=(12,10))
plt.plot(X,yf0,label='f0')
plt.scatter(2,f0(2))
for t in np.linspace(0.1,10,7):
    y = [t*f0(x)+phi(x) for x in X_dom]
    plt.plot(X_dom,y,label='tf0+phi with t = {}'.format(t))
    xstar = min(y)
    plt.scatter(X_dom[y.index(xstar)],xstar)
    plt.annotate('x*(t={})'.format(t)[:10],(X_dom[y.index(xstar)],xstar))
plt.xlabel('x')
plt.legend()
plt.show()
```

(1) Expression for the cross-sectional area of the channel A as a function of a and Î¸ only. 

<img src="img-ass1-cs164-2.png">

The area of the cross-sectional area is equal to the area of the rectangle $(EBCD)_{Area}=bh$ plus the area of the triangle $(ABE)_{Area}=\frac{xh}{2}$.

But, using the defintions of sine and cosine, we get that $h=a sin(\theta)$ and $x= a cos(\theta)$. Furtheremore, we know that $W=h+b+a=3$, and so $b=3-h-a=3-asin(\theta)-a$. 

Therefore,

$$A(a,\theta) = (3-asin(\theta)-a)(asin(\theta)) + \frac{1}{2}acos(\theta)asin(\theta) = $$

$$ 3asin(\theta)-a^2sin^2(\theta)-a^2sin(\theta)+\frac{1}{2}a^2cos(\theta)sin(\theta)$$

Now, the physically-realistic values for $a$ are in $[0,3]$ and for $\theta$ in $0,\frac{\pi}{2}$. This is because $a$ cannot be a negative value since it is the magnitude of the $AB$ segment, and $\theta$ has to be between $0$ and $180$ degrees because otherwise, the drainage would not be able to hold water. 

Given that our domain is $a \in [0,3]$ and $\theta \in [0,\pi]$, we can draw the contour plot of $A(a,\theta)$:

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d 

# %matplotlib notebook

fig = plt.figure(figsize=(8,6))
left, bottom, width, height = 1, 1, 1, 1
ax = fig.add_axes([left, bottom, width, height]) 

#a
x_vals = np.linspace(0, 3, 100)

#theta
y_vals = np.linspace(0, np.pi, 100)
X, Y = np.meshgrid(x_vals, y_vals)


# Z = 0.5*(-2*np.sin(Y)**2+np.sin(Y)*np.cos(Y)-2*np.sin(Y))*X**2+3*np.sin(Y)*X
Z = ((X**2*np.sin(Y)*np.cos(Y))/2) - (X**2*np.sin(Y)**2) + (3*X * np.sin(Y))-(X**2*np.sin(Y))

cp = plt.contour(X, Y, Z,100,cmap='RdBu')
plt.colorbar(cp)

ax.set_title('Contour Plot 2D')
ax.set_xlabel('a (m)')
ax.set_ylabel('theta (radians)')
plt.show()

fig = plt.figure(figsize=(11,8))
ax = fig.add_subplot(111, projection='3d')

ax.contour3D(X,Y,Z,200,cmap='RdBu')
ax.plot_surface(X, Y, Z, cmap='RdBu', rcount=100,alpha=0.5)
plt.colorbar(cp)
ax.set_title('Contour Plot 3D')
ax.set_xlabel('a (m)')
ax.set_ylabel('theta (radians)')
ax.set_zlabel('f(a,theta)')
plt.show()
```

subject to

$W(a,b,\theta) = a+b+asin(\theta) - 3 = 0$

$L(a,b,\theta,\lambda) = absin(\theta)+a^2sin(\theta)cos(\theta) - \lambda(a+b+asin(\theta) - 3)$

By taking the gradient, we get the following system of equations:

(i) $\frac{\partial L}{\partial a} = bsin(\theta)+acos(\theta)sin(\theta)-\lambda-\lambda sin(\theta) = 0 $

(ii) $\frac{\partial L}{\partial b} = asin(\theta)-\lambda = 0 $

(iii) $\frac{\partial L}{\partial \theta} = abcos(\theta)+\frac{1}{2}a^2(cos^2(\theta)-sin^2(\theta))-\lambda a cos(\theta) = 0$

(iv) $\frac{\partial L}{\partial \lambda} = a+b+asin(\theta)-3 = 0$

From this system of equations we obtain the following possible set of solutions:

```python
import numpy as np

#x=a
#y=theta
def line_search(alpha,beta,x,grad,s_init=1):
    v = grad/np.linalg.norm(grad)
    s = s_init
    d = np.dot(grad.T,v)
    while f(x[0]+s*v[0],x[1]+s*v[1]) < f(x[0],x[1])+s*alpha*d:
        s*=beta
    return s

# f(x0) < f(x0 - t*np.array([dfx1(x0), dfx2(x0)])) + alpha * t * np.dot(np.array([dfx1(x0), dfx2(x0)]), np.array([dfx1(x0), dfx2(x0)]))
    
def f(x,y):
    return ((x**2*np.sin(y)*np.cos(y))/2) - (x**2*np.sin(y)**2) + (3*x * np.sin(y))-(x**2*np.sin(y))

def gradf(x,y):
    partial_x= np.sin(y)*(-2*x*np.sin(y)+x*np.cos(y)-2*x+3)
    partial_y= (1/2)*x*(-x*np.sin(y)**2+x*np.cos(y)**2-2*np.cos(y)*(2*x*np.sin(y)+x-3))
    return np.array([partial_x, partial_y])

X0 = np.array([1.5,0]) # Starting state

X = X0
step = 0.01

evolution_Xx = [X0[0]]
evolution_Xy = [X0[1]]
# for i in range(0,1000):
#     X_new = X + step*gradf(X[0], X[1])
# #     print(X)
#     if np.linalg.norm(np.array(X) - np.array(X_new)) <= 1e-6:
#         break
#     X = X_new

X_old = np.array([100,100])
k = 0
while np.linalg.norm(np.array(X) - np.array(X_old)) > 1e-6:
    step = line_search(0.5,0.5,X,gradf(X[0],X[1]),1)
    X_new = X + step*gradf(X[0], X[1])/np.linalg.norm(gradf(X[0], X[1]))
    X_old = X
    X = X_new
    evolution_Xx.append(X[0])
    evolution_Xy.append(X[1])
    k+=1

ev_scatter_Xx = [evolution_Xx[i] for i in range(1,len(evolution_Xx),1)]
ev_scatter_Xy = [evolution_Xy[i] for i in range(1,len(evolution_Xy),1)]

# fig = plt.figure(figsize=(10,8))
# ax = fig.add_subplot(111, projection='3d')
# #a
# x_vals = np.linspace(0, 3, 100)

# #theta
# y_vals = np.linspace(0, np.pi, 100)
# X, Y = np.meshgrid(x_vals, y_vals)
# ax.contour3D(X,Y,Z,200)
# ax.plot(evolution_Xx,evolution_Xy,[f(evolution_Xx[i],evolution_Xy[i]) for i in range(len(evolution_Xy))],linewidth=2,color='black')
# # ax2.plot_surface(X, Y, Z, cmap='RdBu', rcount=100)
# # plt.colorbar(cp)
# ax.set_xlabel('x')
# ax.set_ylabel('y')
# ax.set_zlabel('z')
# plt.show()
# print('Gradient Ascent converged in {} steps'.format(k))

#####
fig = plt.figure(figsize=(10,7))

left, bottom, width, height = 1, 1, 1, 1
ax = fig.add_axes([left, bottom, width, height]) 

X, Y = np.meshgrid(x_vals, y_vals)
cp = plt.contour(X, Y, Z,100,cmap='RdBu')
plt.colorbar(cp)
plt.plot(evolution_Xx,evolution_Xy, label = 'Gradient descent',color='black')
plt.scatter(ev_scatter_Xx,ev_scatter_Xy,s=30,color='black')
plt.legend()
ax.set_title('Contour Plot')
ax.set_xlabel('a (m)')
ax.set_ylabel('theta (radians)')
plt.show()
print('Gradient Ascent converged in {} steps'.format(k))
```
